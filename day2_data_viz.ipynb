```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# 1) View the first 5 rows
display(df.head())

# 2) Dataset info
df.info()

# 3) Summary statistics
df.describe(include='all')
```

Numeric Colums: PassengerId, Survived, pclass, Age, SibSp, Parch, Fare
Categorical columns: Name, Sex, Ticket, Embarked, Cabin
Missing values: Cabin

```python
plt.hist(df['Age'].dropna(), bins=20)
plt.title("Age Distribution on Titanic")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Most people were aged b/w 15-35

```python
survival_counts = df['Survived'].value_counts().sort_index()
survival_counts.plot(kind='bar')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.title("Titanic Survival Counts")
plt.ylabel("Number of Passengers")
plt.show()
```

Survival rate appears to be around 38%
